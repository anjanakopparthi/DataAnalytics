<h3>Global variables</h3>
Note that global variables (such as those in the first cell) aren't used in functions. Note the difference between the two functions below.

```python
def f():
    x=4
f()
print(x)
```

```python
def g():
    global x
    x=4
g()
print(x)
```

In f, a local variable named x is created, and the global variable is unchanged. In g, we specify that we want to use the global variable x.

<h2>Sets in Python</h2>

**Sets** are *unordered* collections of *unique* items. This means a value appears in a set at most once. They are delimited in Python with {}

```python
s={3,4,5}
print(4 in s)
```

Sets are implemented in Python using hash functions, so they can be quite fast. Note how long it takes to find the "last" item in the set, vs the last item in the list. Also note the loop over the set is running 10,000 times vs. 1,000 times for the list!

```python
import time
s={i for i in range(100000)}
l=[i for i in range(100000)]
count=0
startTime=time.time()
for j in range(10000):
    if 99999 in s:
        count+=1
timeElapsed=time.time()-startTime
print(count,timeElapsed, "seconds")
count=0
startTime=time.time()
for j in range(1000):
    if 99999 in l:
        count+=1
timeElapsed=time.time()-startTime
print(count,timeElapsed, "seconds")
```

Sets have the operators you might expect from having taken a discrete math course (union, intersection, issubset, difference, isdisjoint, symmetric_difference)

```python
y={3,4,5,9}
z={3,2,6,9}
print(z.intersection(y))
```

<h2>Dictionaries</h2>

**Dictionaries** are referred to as hash tables or hash maps in other lanuages. They allow you to look up an item quickly with a key.

```python
sportsTeams = { 'Texas A&M':'Aggies', 'University of Texas':'Longhorns', 'Texas Tech':'Red Raiders', 'Texas Christian':'Horned Frogs' }
print(sportsTeams['Texas A&M'])
print("**now lets print them all**")
for key in sportsTeams:
    print(key,":",sportsTeams[key])
print("**dictionary lookup**")
if not "St Mary's" in sportsTeams:
    print("I don't think they play football anymore")
```

<h2>Classes</h2>

Python supports object-oriented classes. As with other languages, classes can have attributes and methods, and can be extended.

```python
class Polygon(object):
    def __init__(self, numSides):
        self.numSides = numSides

class Square(Polygon):
    def __init__(self, sideLength):
        super().__init__(4) # call overridden init!
        self.sideLength = sideLength
    def area(self):
        return self.sideLength*self.sideLength

s=Square(3)
print(s.area())
t=Square(3)
print(s==t)
```

<h4>Storing objects in Dictionaries, Sets, etc.</h4>

If you want sensible things to happen when you store objects in data structures, define these additional methods:

```python
class Polygon(object):
    def __init__(self, numSides):
        self.numSides = numSides
    def getHashables(self):
        return (self.numSides, ) # return a tuple of hashables
    def __hash__(self):
        return hash(self.getHashables())
    def __eq__(self, other):
        return (isinstance(other, Polygon) and (self.numSides == other.numSides))
    
class Square(Polygon):
    def __init__(self, sideLength):
        super().__init__(4) # call overridden init!
        self.sideLength = sideLength
    def area(self):
        return self.sideLength*self.sideLength
    def getHashables(self):
        return (self.selfLength, ) # return a tuple of hashables
    def __hash__(self):
        return hash(self.getHashables())
    def __eq__(self, other):
        return (isinstance(other, Square) and (self.numSides == other.numSides) and (self.sideLength == other.sideLength))
    
s=Square(3)
t=Square(3)
print(s==t)
```
