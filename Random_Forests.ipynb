```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.keys()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

# set up the figure
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    # label the image with the target value
    ax.text(0, 7, str(digits.target[i]))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

Xtrain, Xtest, ytrain, ytest = train_test_split(digits.data, digits.target,
                                                random_state=0)
model = RandomForestClassifier(n_estimators=1000)
model.fit(Xtrain, ytrain)
ypred = model.predict(Xtest)
```

<h3>Classification report</h3>


* precision - what percentage labeled with this category are correct? (true_pos/(true_pos+false_pos))
* recall - what percentage of this category is recognized correctly? (true_pos/(true_pos+false_neg))
* f1 is harmonic mean of precision and recall (2/((1/precision)+(1/recall)))
* support is number of occurrences

From [Wikipedia](https://en.wikipedia.org/wiki/Precision_and_recall)
"Suppose a computer program for recognizing dogs in photographs identifies 8 dogs in a picture containing 12
dogs and 10 cats. Of the 8 identified as dogs, 5 actually are dogs (true positives), 
while the rest are cats (false positives). The program's precision is 5/8 while its recall is 5/12."

```python
from sklearn import metrics
print(metrics.classification_report(ytest, ypred))
```
