<h3>Creating a Tensor (PyTorch matrix)</h3>
Below, we will create an unintialized Tensor. Note it has garbage values (unlike Java that initializes to 0). This saves time, but you need to be careful.

```python
import torch

x = torch.empty(4,2)
print(x)
```

This will initialize with all zeros, using the torch long datatype (a 64 bit integer). We might also use torch.double for a 64 bit float

```python
x=torch.zeros(4,2,dtype=torch.long)
print(x)
```

Or we could throw in random numbers

```python
x=torch.rand(4,2)
print(x)
```

We can also create tensors from numpy arrays. Note how the numpy array and the tensor share the same memory.

```python
import numpy as np
y=np.array([[1, 2, 3], [4, 5, 6]])
print(y)
z=torch.from_numpy(y)
print(z)
y[1,2]=23
print(z)
```

Now let's copy the Tensor over to the GPU

```python
if torch.cuda.is_available():
    device = torch.device("cuda")
    onGpu=z.to(device)
    print(onGpu)
else:
    print("no CUDA for you!")
```
