### Task 1: 
Load the Abalonedataset from https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/into your Jupyter notebook

#### Approach
1. Import pandas to use read_csv
2. load the abalone data from the following url:  https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data

```python
import pandas as pd
import numpy as ny
from matplotlib import pyplot as plt 
```

```python
column_names = ['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings']
abalone_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data", names = column_names)
```

```python
# To understand the data, use the following commands
# abalone_data.head() 
# abalone_data.shape
# abalone_data.info()
# Columns name are not available in the loaded data. Get the column name from Attribute information from abalone.names.
# Check the data after adding column names
# abalone_data.info()
```

### Task 2: 
For each of length, diameter, height, rings, have your notebook output the min, max, mean, and standarddeviation 
### Approach:
Use min, max, mean and std functions on all attributes Length, Diameter, Height and Rings 

```python
attributes = ['Length','Diameter','Height','Rings']
Stats      = ['min','max','mean','standard deviation']
arr        = ny.zeros((4,4))

for i in range(4):
    arr[0,i] = abalone_data[attributes[i]].min()
    arr[1,i] = abalone_data[attributes[i]].max()
    arr[2,i] = abalone_data[attributes[i]].mean()
    arr[3,i] = abalone_data[attributes[i]].std()
```

```python
Result = pd.DataFrame(arr, columns=attributes, index=Stats)
```

```python
# print(Results)
# To round the output to 3 decimal points
print(round(Result,3))
```

### End of Task 2

### Task 3
Have your notebook create histograms for the length, diameter, height and weight of male, female and infant abalone (12 histograms)

```python
Sex_data = ['M','F','I']
Sex_Attr = {'M':'Male', 'F':'Female','I':'Infant'}
hist_attributes = ['Length', 'Diameter', 'Height','Whole weight']

for i in Sex_data:
    for j in hist_attributes:
        abalone_temp = abalone_data[abalone_data['Sex'] == i] [j]
        #plt.style.use('fivethirtyeight')
        plt.hist(abalone_temp, edgecolor = 'black')
        # To highlight the outliers use 'log = True'
        mean = abalone_temp.mean()
        red = '#fc4f30'
        plt.axvline(mean,color = red, label = 'Mean ' + j)
        plt.legend()
        plt.title('Histogram for '+j+ ' of '+ Sex_Attr[i])
        plt.xlabel(j)
        plt.ylabel('Population')
        plt.tight_layout()
        plt.show()
```

### End of Task 3

### Observation for Task 3
Some outliers are not clearing visible in the above plots, to highlight the outliers we can plot on a logarthimic scale

```python
Sex_data = ['M','F','I']
Sex_Attr = {'M':'Male', 'F':'Female','I':'Infant'}
hist_attributes = ['Length', 'Diameter', 'Height','Whole weight']

for i in Sex_data:
    for j in hist_attributes:
        abalone_temp = abalone_data[abalone_data['Sex'] == i] [j]
        #plt.style.use('fivethirtyeight')
        plt.hist(abalone_temp, edgecolor = 'black', log = True)
        # To highlight the outliers use 'log = True'
        mean = abalone_temp.mean()
        red = '#fc4f30'
        plt.axvline(mean,color = red, label = 'Mean ' + j)
        plt.legend()
        plt.title('Histogram for '+j+ ' of '+ Sex_Attr[i])
        plt.xlabel(j)
        plt.ylabel('Population (log axis)')
        plt.tight_layout()
        plt.show()
```
